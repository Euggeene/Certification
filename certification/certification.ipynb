```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('kc_house_data.csv' , sep=',')
df.head()
```

# Условие 1: Задача 1
Постройте график.
Назовите график.

```python
plt.figure(figsize=(8,6))
plt.hist(df['price'], bins=50)
plt.title('Стоимость недвижимости')
plt.xlim(0, 4000000)
plt.xlabel('Цена $')
plt.ylabel('Кол-во домов');
```

Вывод:

```python
plt.figure(figsize=(8,6))
plt.hist(df['sqft_living'], bins=30)
plt.title('Распределение жилой квадратуры')
plt.xlim(0, 10000)
plt.xlabel('Квадратные футы')
plt.ylabel('Кол-во домов');
```

Вывод:
Жилая квадратная площадь домов составляет в среднем 2тыс.кв.футов

```python
plt.figure(figsize=(8,6))
plt.title('Распределение года постройки')
plt.hist(df['yr_built'], bins=50)
plt.xlabel('Год постройки')
plt.ylabel('Кол-во домов');
```

```python
plt.figure(figsize=(8,6))
data = df['yr_built'].value_counts()
plt.bar(data.index, data.values)
plt.xlabel('Год постройки')
plt.ylabel('Кол-во домов')
plt.xticks(rotation=90);
```

Вывод:
Кризисы строительтва были в середине 30-хб начале 70-х. Пики были в конце 60-хб и 70-х, а также в середине нулевых и десятых годов 21 века.

# Условие 2: Задача 2.
2.1. Изучите распределение домов от наличия вида на набережную.
Постройте график.
Сделайте выводы.
2.2. Изучите распределение этажей домов.
2.3. Изучите распределение состояния домов.

```python
data = df['waterfront'].value_counts()
plt.figure(figsize=(8,6))
plt.title('Распределение домов от наличия вида на набережную')
sns.barplot(x=data.index, y=data.values)
plt.xlabel('Вид на набережную')
plt.ylabel('Кол-во домов');
```

```python
data2 = df['waterfront'].value_counts()
plt.figure(figsize=(8,6))
plt.title('Распределение домов от наличия вида на набережную')
plt.pie(data2.values, autopct="%1.1f%%")
plt.legend(['Без вида на набережную', 'С видом на набережную']);
```

Вывод:
Меньше 1% домов имеет вид на набережную.

```python
data = df['floors'].value_counts()
plt.figure(figsize=(8,6))
plt.title('Распределение этажей')
sns.barplot(x=data.index, y=data.values)
plt.xlabel('Кол-во этажей')
plt.ylabel('Кол-во домов');
```

```python
data2 = df['floors'].value_counts()
plt.figure(figsize=(8,6))
plt.title('Распределение этажей')
plt.pie(data2.values, autopct="%1.1f%%", labels=data.index);
```

Вывод: 
Наиболее распространены одноэтажные и двухэтажные дома. Меньше всего домов с 3.5 этажами.

```python
data = df['condition'].value_counts()
plt.figure(figsize=(8,6))
plt.title('Состояние домов')
sns.barplot(x=data.index, y=data.values)
plt.xlabel('Оценка')
plt.ylabel('Кол-во домов');
```

```python
data2 = df['condition'].value_counts()
plt.figure(figsize=(8,6))
plt.title('Состояние домов')
plt.pie(data2.values, autopct="%1.1f%%", labels=data.index);
```

Вывод: 
Менее 1% домов находятся в плохом состоянии. Около 2/3 домов в удовлетворительном состоянии.

# Условие 3: Задача 3.
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corr_matrix = df.corr(numeric_only=True)
corr_matrix
```

```python
sns.heatmap(corr_matrix, cmap='coolwarm');
```

```python
plt.figure(figsize=(12,10))
sns.boxplot(x=df['condition'].astype('str'), y=df['price'], whis=1.5);
plt.xlabel('Состояние')
plt.ylabel('Цена')
plt.title('Распределение цен в зависимости от состояния дома');
```

```python
plt.figure(figsize=(12, 10))
sns.boxplot(x=df['grade'].astype('str'), y=df['price'], whis=1.0)
plt.xlabel('Класс')
plt.ylabel('Цена')
plt.title('Распределение цен в зависимости от класса дома');
```

```python
plt.figure(figsize = (12,10))
sns.boxplot(x = df['price'], y = df['floors'].astype('str'), whis = 1.5)
plt.xlabel('Цена')
plt.ylabel('Этажность')
plt.title('Распределение цен в зависимости от этажности дома');
```

```python
plt.figure(figsize=(12, 10))
sns.boxplot(x=df['price'], y=df['waterfront'].astype('str'), whis=1.5, hue=df['grade'])

plt.title('Распределение цен по береговой линии')
plt.xlabel('Цена')
plt.ylabel('Наличие набережной');
```
